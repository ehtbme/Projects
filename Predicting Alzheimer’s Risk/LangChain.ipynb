```python
from langchain import PromptTemplate, LLMChain
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load BioGPT model from HuggingFace
tokenizer = AutoTokenizer.from_pretrained("microsoft/BioGPT")
model = AutoModelForCausalLM.from_pretrained("microsoft/BioGPT")

# Build an HF pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100)

# Wrap it as a LangChain LLM
llm = HuggingFacePipeline(pipeline=pipe)

# Define a prompt template
template = """You are a clinical AI model. Summarize the key symptoms and suggest potential risks:
{clinical_notes}
"""
prompt = PromptTemplate(template=template, input_variables=["clinical_notes"])

# Create a chain
chain = LLMChain(prompt=prompt, llm=llm)

# Example input
clinical_text = """The patient shows mild memory loss, confusion, and disorientation during late evenings."""
output = chain.run(clinical_notes=clinical_text)
print(output)
```
